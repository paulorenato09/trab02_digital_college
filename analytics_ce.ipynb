```python
import unicodedata

# Função para remover acentos e caracteres especiais
def remove_accent(text):
    text = unicodedata.normalize('NFKD', text)
    return text.encode('ASCII', 'ignore').decode('ASCII')

df_rg_adm['titulo'] = df_rg_adm['titulo'].apply(remove_accent)

print(df_rg_adm)
```

```python
# Dando replace nas variáveis onde ocorreu erros de digitação ou formatação:

df_rg_adm.replace('SERTAO DOS CRATEAsS', 'SERTAO DOS CRATEUS', inplace=True)
df_rg_adm.replace('MACIAO DO BATURITA', 'MACICO DE BATURITE', inplace=True)

df_rg_adm
```

# Conclusão:
<font size="3">
            Uma correlação de 0.27 entre as colunas valinicial e valempenhado indica uma correlação positiva fraca. 
</font>

```python
def gerar_grafico(df_filtrado, df_val):
    # Criando uma figura para armazenar os gráficos
    fig = plt.figure(figsize=(18, 16))  # Aumentar o tamanho da figura para acomodar mais gráficos

    # Gráfico de barras para a coluna 'iduso'
    plt.subplot(4, 1, 1)
    sns.countplot(data=df_filtrado, x='iduso', palette='Set2')
    plt.title('Distribuição de iduso')
    plt.xlabel('iduso')
    plt.ylabel('Frequência')

    # Gráfico de barras para a coluna 'ug'
    plt.subplot(4, 1, 2)
    sns.countplot(data=df_filtrado, x='ug', palette='Set1')
    plt.title('Distribuição de UG')
    plt.xlabel('UG')
    plt.ylabel('Frequência')

    # Gráfico de barras para a coluna 'localizadorgasto'
    plt.subplot(4, 1, 3)
    sns.countplot(data=df_filtrado, x='localizadorgasto', palette='Set3')
    plt.title('Distribuição de Localizador de Gasto')
    plt.xlabel('Localizador de Gasto')
    plt.ylabel('Frequência')

    # Gráfico de correlação
    plt.subplot(4, 1, 4)
    corr = df_val[['valinicial', 'valempenhado']].corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlação entre Valinicial e Valempenhado')

    # Ajustando o layout para evitar sobreposição
    plt.tight_layout()

    return fig  # Retorna a figura gerada

# Exemplo de uso
grafico = gerar_grafico(df_filtrado, df_val)
```

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
```

```python
def enviar_email(df_filtrado, grafico):
    # Configurações do e-mail
    remetente = 'paulorenatolima22@gmail.com'
    senha = 'gzsp eyer vedm lyir'
    destinatario = 'paulorenatolima22@gmail.com'

    # Configurações do servidor SMTP (usando o Gmail neste exemplo)
    servidor_smtp = 'smtp.gmail.com'
    porta_smtp = 587

    # Criando o corpo do e-mail
    mensagem = MIMEMultipart()
    mensagem['From'] = remetente
    mensagem['To'] = destinatario
    mensagem['Subject'] = 'Relatório De Faturamento do estado do Ceará'

    # Adicionando o corpo do e-mail
    corpo_email = 'Segue gráficos que foram resultado da análise.'
    mensagem.attach(MIMEText(corpo_email, 'plain'))

    # Salvando o gráfico como uma imagem
    grafico.savefig('grafico.png')  # Salva a imagem do gráfico
    # Não fechar o gráfico ainda, apenas após o envio do e-mail

    # Adicionando o gráfico como anexo
    with open('grafico.png', 'rb') as img:
        attachment = MIMEApplication(img.read(), _subtype="png")
        attachment.add_header('Content-Disposition', 'attachment', filename='grafico.png')
        mensagem.attach(attachment)

    # Conectando ao servidor SMTP e enviando o e-mail
    with smtplib.SMTP(servidor_smtp, porta_smtp) as server:
        server.starttls()  # Segurança
        server.login(remetente, senha)  # Login no servidor
        server.sendmail(remetente, destinatario, mensagem.as_string())  # Envio do e-mail

    # Fechando o gráfico após o envio do e-mail
    # grafico.close()

# Exemplo de uso (supondo que df_filtrado e grafico já estão definidos e criados)
enviar_email(df_filtrado, grafico)
```
